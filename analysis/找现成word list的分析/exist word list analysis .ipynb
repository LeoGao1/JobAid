```python
import numpy as np
import pandas as pd
import PyPDF2 # a package of pdf reader 
from bs4 import BeautifulSoup
import requests
import operator
import re
```

1.找emotional word list


我是在网上找到了一个pdf文档，里面包含了一些关于emotion的词语

```python

#这一部分主要是从pdf得到一个word list
emotional_pdf = 'Practices-FeelingsSensations.pdf'
pdfFileObj = open(emotional_pdf, 'rb')
pdfReader = PyPDF2.PdfFileReader(pdfFileObj)
pageObj = pdfReader.getPage(0)
text = pageObj.extractText().split('\n')
text = [i for i in text if(len(i)>0)]
text.remove('Feelings List')
text.remove('OpenCalmCentered')
text.remove( 'HopelessLonelySorrow')
text.remove('HumbledLuckyMoved')
text.remove('LooseNauseous Numb Pain')
text.remove('SpaciousSparkly')
text.remove('Burned outCranky')
text.remove('ColdConstricted')
text.remove('DullElectricEmpty')
text.remove('PricklyQueasy')

emotional_text = text
emotional_text = [i.lower() for i in emotional_text]
emotional_text
```

2.找data science 的对应词汇


在网上找到了一个包含相关词汇的网站（http://www.datascienceglossary.org/#algorithm）


爬了下来转换成一个word list

```python
page1 = requests.get('http://www.datascienceglossary.org/#algorithm')
page1.ok
```

```python
soup1 = BeautifulSoup(page1.content, 'html.parser')
```

```python
dsc_text = []
for i in soup1.find_all('dfn'):
    dsc_text.append(i.get_text())
    
dsc_text = [i.lower() for i in dsc_text]
dsc_text
```

3. 找到programming 的相关词汇


```python
page2 = requests.get('https://www.computerhope.com/jargon/program.htm')
page2.ok
```

```python
soup2 = BeautifulSoup(page2.content, 'html.parser')
```

```python
prog_text = []
for i in soup2.find_all('a'):
    prog_text.append(i.get_text())
    
prog_text = prog_text[66:-20]
prog_text = [i.lower() for i in prog_text if (len(i)>0)]
prog_text
```

之后有更多list可以再加




接下来看这些词在data里面出现的频率


```python
df = pd.read_csv('Data.csv')
df.head()
```

```python
def clean_questions(question):
        question = re.sub(r'[^\w\s]','',question)
        return (question.lower().replace('&', 'and'))
```

```python
df['Question'] = df['Question'].apply(clean_questions)
```

```python
df.head()
```

```python
emotion_dic = {}
for i in emotional_text:
    emotion_dic[i] = 0
    
dsc_dic = {}
for i in dsc_text:
    dsc_dic[i] = 0
    
prog_dic = {}
for i in prog_text:
    prog_dic[i] = 0

for i in list(df['Question']):
    word_list = i.split(' ')
    for w in range(len(word_list)):
        word_list[w] = word_list[w].lower()
        
    for em in emotional_text:
        if(em in word_list):
            emotion_dic[em] += 1
    
    for d in dsc_text:
        if(d in word_list):
            dsc_dic[d] += 1
            
    for p in prog_text:
        if(p in word_list):
            prog_dic[p] += 1
            
```
