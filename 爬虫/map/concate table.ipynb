```python
import time
import json
from bs4 import BeautifulSoup
import requests
import pandas as pd
import numpy as np 
```

```python
list1 = ["data analyst",
      "data scientist",
       "data engineer",
        "Consultant",
        "Accountant",
        "Graphic Designer",
        "UX Designer",
        "UI Designer",
        "Art Director",
        "Mental health counselor"]
       
```

```python
names= [i.lower().replace(' ','_')+'_locations.csv' for i in list1]
names
```

```python
df1 = pd.read_csv(names[0])
df2 = pd.read_csv(names[1])
df3 = pd.read_csv(names[2])
df = [df1,df2,df3]
```

```python

result = pd.concat(df,axis = 0,ignore_index = True).groupby('state').sum().sort_values(by='count',ascending = False)
result
```
